```python
import pandas as pd
```

```python
harvard = pd.read_csv("Harvard.csv")
penn = pd.read_csv("Penn.csv")
```

```python
harvard.columns
```

```python
harvard = harvard.drop(['No.', '#', 'C', 'Personal Pronouns'], axis = 1)
```

```python
harvard.columns
```

```python
harvard
```

```python
harvard = harvard.fillna(" ")
harvard['year'] = harvard['year'].astype(str)
harvard['Year'] = harvard['Year'].astype(str)
harvard['Name'] = harvard['Name'] + harvard['Full Name']
harvard = harvard.drop('Full Name', axis = 1)
harvard['Academic Year'] = harvard['Yr.'] + harvard['Academic Year']
harvard = harvard.drop('Yr.', axis = 1)
harvard['Pos.'] = harvard['Pos.'] + harvard['Position'] + harvard['Pos'] + harvard['Event']
harvard = harvard.drop(['Position', 'Pos', 'Event'], axis = 1)
harvard['year'] = harvard['year'] + harvard['Year']
harvard = harvard.drop('Year', axis = 1)
harvard['sport'] = harvard['sport'] + harvard['Sport']
harvard = harvard.drop('Sport', axis = 1)
```

```python
harvard
```

```python
import numpy as np
harvard["Hometown / High School"] = harvard["Hometown / High School"].replace(r'^\s*$', np.nan, regex=True)
```

```python
harvard
```

```python
harvard["Hometown / High School"].fillna("Unknown / Unknown", inplace = True)
harvard[['hmtwn', 'hs', 'bs']] = harvard['Hometown / High School'].str.split('/', expand = True)
# harvard['hmtwn'] = harvard['Hometown / High School'].str.split('/')
# havard['hs'] = harvard['Hometown / High School'].str.split('/')
harvard = harvard.drop(['Hometown / High School', 'bs'], axis = 1)
harvard['hmtwn'] = harvard['hmtwn'].replace('Unknown ', ' ')
harvard['hs'] = harvard['hs'].replace(' Unknown', ' ')
```

```python
harvard
```

```python
harvard["Hometown"] = harvard["Hometown"] + harvard["hmtwn"]
harvard = harvard.drop("hmtwn", axis = 1)
harvard["High School"] = harvard["High School"] + harvard["hs"]
harvard = harvard.drop("hs", axis = 1)
harvard
```

```python
harvard["Wt."] = harvard["Wt."].replace(' ', pd.NA)
harvard["Ht."] = harvard["Ht."].replace(' ', pd.NA)
harvard
```

```python
harvard['college'] = "Harvard"
harvard
```

```python
harvard['year'] = harvard['year'].astype('float').astype(int)
harvard
```

```python
sport = harvard.pop('sport')
harvard.insert(1, 'Sport', sport)
college = harvard.pop('college')
harvard.insert(2, 'College', college)
position = harvard.pop('Pos.')
harvard.insert(3, 'Position', position)
height = harvard.pop('Ht.')
harvard.insert(4, 'Height', height)
weight = harvard.pop('Wt.')
harvard.insert(5, 'Weight', weight)
acayear = harvard.pop('Academic Year')
harvard.insert(6, 'Academic Year', acayear)
ht = harvard.pop('Hometown')
harvard.insert(7, 'Hometown', ht)
hs = harvard.pop('High School')
harvard.insert(8, 'High School', hs)
year = harvard.pop('year')
harvard.insert(9, 'Year', year)
```
