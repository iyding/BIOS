```python
import os
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium import webdriver
from selenium.webdriver.common.by import By
import re
import pandas as pd
options = Options()
options.headless = True
service = Service(executable_path="chromium.chromedriver")
driver = webdriver.Chrome(options=options, service=service)
from bs4 import BeautifulSoup
import requests
```

```python
driver.get("https://yalebulldogs.com")
links = driver.find_elements("tag name", "a")
```

```python
url = "https://yalebulldogs.com"
href_list = []
url += "/sports"
sports = ["football","basketball","field","volleyball","soccer"]
for link in links:
    try:
        href = link.get_attribute("href")
        href_split_slash = href.split("/")
        href_sports = href_split_slash[-1].split("-")
        href_sport = href_sports[-1]
        href += "/roster"
        if re.match(url,href) and href not in href_list and href_sport in sports:
            href_list.append(href)
    except:
        continue
```

```python
two_year_list = ["2010-11", "2011-12", "2012-13", "2013-14", 
                 "2014-15","2015-16","2016-17","2018-19",
                 "2019-20","2020-21", "2021-22", "2022-23", "2023-24"]
year_list = ["2010", "2011", "2012", "2013", "2014", "2015", "2016", 
             "2017", "2018", "2019", "2020", "2021", "2022", "2023"]
two_year_sports = ['https://yalebulldogs.com/sports/womens-basketball/roster',
                  'https://yalebulldogs.com/sports/womens-track-and-field/roster',
                   'https://yalebulldogs.com/sports/mens-track-and-field/roster',
                   'https://yalebulldogs.com/sports/mens-basketball/roster'
                  'https://yalebulldogs.com/sports/mens-soccer/roster']
one_year_sports = ['https://yalebulldogs.com/sports/womens-soccer/roster',
                  'https://yalebulldogs.com/sports/womens-volleyball/roster',
                  'https://yalebulldogs.com/sports/football/roster']
```

```python
merge_list = []
for href in href_list:
    if href in two_year_sports:
        for year in two_year_list:
            href_year = href + "/" + year
            df = pd.read_html(href_year)[2]
            df["year"] = year
            df["team"] = re.findall("//.*/.*/(.*)/.*$", href)[0].replace("-", " ")
            merge_list.append(df)
    if href in one_year_sports:
        for year in year_list:
            href_year = href + "/" + year
            df = pd.read_html(href_year)[2]
            df["year"] = year
            df["team"] = re.findall("//.*/.*/(.*)/.*$", href)[0].replace("-", " ")
            merge_list.append(df)
full_df = pd.concat(merge_list)
```

```python
#Yale Processing
import pandas as pd
import numpy as np
complete_df["Number"] = complete_df["#"]

df2 = complete_df

zero_list = list(df2[df2["Number"] == 0.0].index)

df2["Name"] = df2["Full Name"].fillna("") + df2["Name"].fillna("")

def number(row):
    if row["Number"] != np.NaN:
        return row["Number"]
    else:
        return row["No."]

df2["Number"] = df2["Number"].fillna(0)
df2["No."] = df2["No."].fillna(0)

df2["Number"] = df2["Number"] + df2["No."]

df2 = df2.drop(columns = ["No.","Events", "Major", "Club","Custom Field 1"])

df2[["Academic Year", "Pos", "Cl.", "Pos."]] = df2[["Academic Year", "Pos", "Cl.", "Pos."]].fillna("")

df2["Academic Year"] = df2["Academic Year"] + df2["Cl."]

df2["Position"] = df2["Pos"] + df2["Pos."]

df2["Hometown / High School / Club"] = df2["Hometown / High School / Club"].fillna("")

df2[["Hometown / High School / Club","Hometown / High School"]] = df2[["Hometown / High School / Club","Hometown / High School"]].fillna("")

df2["Hometown/Highschool"] = df2["Hometown / High School / Club"] + df2["Hometown / High School"]

def split_hometown(row):
    split = row["Hometown/Highschool"].split("/")
    strip = split[0].strip()
    return strip

df2["Hometown"] = df2.apply(split_hometown, axis = 1)

def split_highschool(row):
    split = row["Hometown/Highschool"].split("/")
    strip = split[-1].strip()
    return strip



df2["Highschool"] = df2.apply(split_highschool, axis = 1)

df2 = df2.drop(columns = ["Pos","Hometown / High School","Cl.","Hometown / High School / Club","Hometown/Highschool"])

df2 = df2.drop(columns = "Full Name")

df2["Number"] = df2["Number"].replace(to_replace = 0,value = np.NaN)

df2 = df2.drop(columns = "Pos.")


for idx in zero_list:
    df2["Number"][idx] = 0

df2["Height"] = df2["Ht."]
df2["Team"] = df2["team"]
df2["Year"] = df2["year"]
df2["Weight"] = df2["Wt."]
df2= df2.drop(columns = ["Ht.","team","year","Wt."])

df2["Number"] = df2["Number"].fillna(30000000)


df2["Number"] = df2["Number"].replace(to_replace = 30000000 ,value = np.NaN)

df2["Height"] = df2["Height"].replace(to_replace = "-" ,value = np.NaN)

df2.to_csv("yale.csv")
```

```python
two_year_list = ["2010-11", "2011-12", "2012-13", "2013-14", 
                 "2014-15","2015-16","2016-17","2018-19",
                 "2019-20","2020-21", "2021-22", "2022-23", "2023-24"]
year_list = ["2010", "2011", "2012", "2013", "2014", "2015", "2016", 
             "2017", "2018", "2019", "2020", "2021", "2022", "2023"]
two_year_sports_brown_men = ['https://brownbears.com/sports/mens-basketball/roster'
                  'https://brownbears.com/sports/mens-soccer/roster']
two_year_sports_brown_women = ['https://brownbears.com/sports/womens-basketball/roster']
one_year_sports_brown = ['https://brownbears.com/sports/womens-soccer/roster',
                  'https://brownbears.com/sports/womens-volleyball/roster']
```

```python
#women's soccer and volleyball
merge_list = []
for team in one_year_sports_brown:
    for year in year_list:
        url = team + "/" + year
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
        player_names = []
        player_positions = []
        player_number = []
        player_height = []
        player_year = []
        player_weight = []
        player_home = []
        player_school = []
        player_list = soup.find_all("div", class_ = "s-person-card s-person-card--list flex flex-col overflow-hidden rounded-[10px] border s-person-card--theme-light-theme shadow-level-1 border")
        for player in player_list:
            span_list = player.find_all("span")
            if len(span_list) < 13:
                continue
            player_number.append(span_list[0].text)
            player_names.append(span_list[1].text)
            player_positions.append(span_list[2].text[9:])
            player_year.append(span_list[4].text[14:])
            player_height.append(span_list[6].text[6:])
            player_home.append(span_list[9].text[9:])
            player_school.append(span_list[11].text[12:])
        df_dict = {}
        df_dict["Name"] = player_names
        df_dict["Position"] = player_positions
        df_dict["Number"] = player_number
        df_dict["Height"] = player_height
        df_dict["Academic Year"] = player_year
        df_dict["Hometown"] = player_home
        df_dict["Highschool"] = player_school
        df = pd.DataFrame(df_dict)
        df["Team"] = re.findall("sports/(.*)/roster", url)[0].replace("-", " ")
        df["Weight"] = np.NaN
        df["Year"] = year
        merge_list.append(df)
womens_one_year_df = pd.concat(merge_list)
```
