```python
all_years_players = pd.DataFrame()
sports = ['mens-basketball', 'womens-basketball']
oneyr_sports = ["football", 'womens-volleyball', 'womens-soccer', 'mens-soccer']
track =  'mens-track-and-field' # 2 tables on same page; one for mens, one for womens

years = range(2010, 2025, 1)

for sport in sports:
    for year in years:
        if year == 2024:
            print(year - 1, sport)
            yeardf = pd.read_html(f"https://gocrimson.com/sports/{sport}/roster?view=2")[2]
            yeardf['year'] = year - 1
            yeardf['sport'] = sport
            #print(yeardf.head())
            all_years_players = pd.concat([ yeardf, all_years_players], ignore_index=True)
        else:
            print(year, sport)
            yeardf = pd.read_html(f"https://gocrimson.com/sports/{sport}/roster/{year-1}-{year-2000}?view=2")[2]
            yeardf['year'] = year - 1
            yeardf['sport'] = sport
            #print(yeardf.head())
            all_years_players = pd.concat([ yeardf, all_years_players], ignore_index=True)

```

```python
all_years_players.head()
```

```python
for sport in oneyr_sports:
    for year in years:
            print(year - 1, sport)
            yeardf = pd.read_html(f"https://gocrimson.com/sports/{sport}/roster/{year}?view=2")[2]
            yeardf['year'] = year - 1
            yeardf['sport'] = sport
            all_years_players = pd.concat([all_years_players, yeardf], ignore_index = True)
```

```python
all_years_players
```

```python
for year in range(2016,2025,1):
    mens_track = pd.read_html(f"https://gocrimson.com/sports/mens-track-and-field/roster/{year-1}-{year-2000}?view=2")[3]
    womens_track = pd.read_html(f"https://gocrimson.com/sports/mens-track-and-field/roster/{year-1}-{year-2000}?view=2")[2]
    mens_track["Sport"] = "track-and-field"
    womens_track["Sport"] = "track-and-field"
    mens_track["Year"] = year-1
    womens_track["Year"] = year-1
    print(year)
    
    all_years_players = pd.concat([all_years_players, mens_track, womens_track], ignore_index = True)
```
