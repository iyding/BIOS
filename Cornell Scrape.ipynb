```python
import pandas as pd
```

```python
years = list(range(2010, 2025))
sports = ["football", 'mens-basketball', 'womens-basketball', 'womens-soccer', 'mens-soccer', 'womens-track-and-field',
          'mens-track-and-field']
oneyr_sports = ['womens-volleyball'] 

all_years_players = pd.DataFrame()
for sport in sports:
    for year in years:
        if year == 2024:
            print(year - 1, sport)

            yeardf = pd.read_html(f"https://cornellbigred.com/sports/{sport}/roster?view=2")[2]
            yeardf['year'] = year - 1
            yeardf['Sport'] = sport
            #print(yeardf.head())
            all_years_players = pd.concat([ yeardf, all_years_players], ignore_index=True)
        else:
            print(year, sport)
            yeardf = pd.read_html(f"https://cornellbigred.com/sports/{sport}/roster/{year-1}-{year-2000}?view=2")[2]
            yeardf['year'] = year - 1
            yeardf['Sport'] = sport
            #print(yeardf.head())
            all_years_players = pd.concat([all_years_players, yeardf], ignore_index=True)

for sport in oneyr_sports:
    
    for year in years:
            print(year - 1, sport)
            yeardf = pd.read_html(f"https://cornellbigred.com/sports/{sport}/roster/{year}?view=2")[2]
            yeardf['year'] = year - 1
            yeardf['Sport'] = 'volleyball'
            all_years_players = pd.concat([all_years_players, yeardf], ignore_index = True)
```

```python
all_years_players.columns.values
```

```python
all_years_players['Name'] = (all_years_players['Name'].fillna(" ") + all_years_players['Full Name'].fillna(" ")).str.strip()
```

```python
hometown = all_years_players['Hometown / High School'].str.split("/", expand=True)
hometown
all_years_players['Hometown_2'] = hometown[0]
all_years_players['High_school_2'] = hometown[1]
```

```python
hometown = all_years_players['Hometown / Last School'].str.split("/", expand=True)
hometown
all_years_players['Hometown_3'] = hometown[0]
all_years_players['High_school_3'] = hometown[1]
```

```python
all_years_players[['Hometown_4', 'High School_4']] = all_years_players['Hometown/High School'].str.split("/", expand=True)
```

```python
all_years_players['Height'] = (all_years_players['Ht.'].fillna(" ") + all_years_players['Height'].fillna(" ")).str.strip()
all_years_players['Academic Year'] = (all_years_players['Cl.'].fillna(" ") + all_years_players['Year'].fillna(" ")).str.strip()
all_years_players['Hometown'] = (all_years_players['Hometown_2'].fillna(" ") + all_years_players['Hometown_3'].fillna(" ") + all_years_players['Hometown_4'].fillna(" ")).str.strip()
all_years_players['High School'] = (all_years_players['High_school_2'].fillna(" ") + all_years_players['High_school_3'].fillna(" ") + all_years_players['High School_4'].fillna(" ")).str.strip()
all_years_players['Position'] = (all_years_players['Pos.'].fillna(" ") + all_years_players['Events'].fillna(" ")).str.strip()
all_years_players['Weight'] = all_years_players['Wt.']
all_years_players["College"] = 'Cornell'
```

```python
cols =("""Name
Sport
College
Position
Height
Weight
Academic Year
Hometown
High School
year""".split("\n"))

```

```python
df = all_years_players.loc[:, cols]
```

```python
df.rename(columns={'year': 'Year'}, inplace=True)
```
