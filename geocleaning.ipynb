```python
import pandas as pd
```

```python
df = pd.read_csv("harvard_penn_complete.csv")
```

```python
df = df.drop("Unnamed: 0", axis = 1)
```

```python
def stand_city_name(city):
    us_state_abbrev = {
        'ALA.': 'Alabama',
        'AK.': 'Alaska',
        'ARIZ.': 'Arizona',
        'AR.': 'Arkansas',
        'CALIF.': 'California',
        'COLO.': 'Colorado',
        'CONN.': 'Connecticut',
        'D.C.': 'District of Columbia',
        'DE.': 'Delaware',
        'FLA.': 'Florida',
        'GA.': 'Georgia',
        'HI.': 'Hawaii',
        'ID.': 'Idaho',
        'ILL.': 'Illinois',
        'IND.': 'Indiana',
        'IA.': 'Iowa',
        'KAN.': 'Kansas',
        'KY.': 'Kentucky',
        'LA.': 'Louisiana',
        'ME.': 'Maine',
        'MD.': 'Maryland',
        'MASS.': 'Massachusetts',
        'MICH.': 'Michigan',
        'MINN.': 'Minnesota',
        'MS.': 'Mississippi',
        'MO.': 'Missouri',
        'MT.': 'Montana',
        'NE.': 'Nebraska',
        'NEV.': 'Nevada',
        'N.H.': 'New Hampshire',
        'N.J.': 'New Jersey',
        'N.M.': 'New Mexico',
        'N.Y.': 'New York',
        'N.C.': 'North Carolina',
        'N.D.': 'North Dakota',
        'OH.': 'Ohio',
        'OKLA.': 'Oklahoma',
        'ORE.': 'Oregon',
        'PA.': 'Pennsylvania',
        'R.I.': 'Rhode Island',
        'S.C.': 'South Carolina',
        'S.D.': 'South Dakota',
        'TENN.': 'Tennessee',
        'TX.': 'Texas',
        'UT.': 'Utah',
        'VT.': 'Vermont',
        'VA.': 'Virginia',
        'WASH.': 'Washington',
        'W.V.': 'West Virginia',
        'WIS.': 'Wisconsin',
        'WY.': 'Wyoming',
        'del.': 'Delaware',
        'fl.': 'Florida',
        'miss.': 'Mississippi',
        'calif': 'California',
        'fl.': 'Florida',
        'ill.': 'Illinois',
        'tex.':'Texas',
        'il':'Illinois',
        'ks':'Kansas',
        'wa.':'Washington',
        'wva.':'West Virginia',
        'or':'Oregon',
        'wash':'Washington',
        'il.':'Illinois',
        'ga':'Georgia',
        'oh':'Ohio',
        'ct.':'Connecticut',
        'or.':'Oregon',
        'ny':'New York',
        'penn.':'Pennsylvania',
        'wisc.':'Wisconsin',
        'neb.':'Nebraska',
        'w.va.':'West Virginia',
        'mont.':'Montana',
        'ark.': 'Arkansas',
        'del.': 'Delaware'  
    }
    if city is None or not isinstance(city, str) or ',' not in city:
        return None

    city_parts  = city.split(", ")
    city_name = city_parts[0].strip().lower()
    
    if len(city_parts) < 2:
        return city_name
    
    state_abbr = city_parts[1].strip().lower()
    state_name = us_state_abbrev.get(state_abbr.upper(), state_abbr)
    
    return f"{city_name}, {state_name.lower()}"


df['std_city_name'] = df['Hometown'].apply(stand_city_name)
    
```

```python
df[['city', 'state']] = df['std_city_name'].str.split(',', 1, expand=True)
df['state'] = df['state'].str.lstrip()
```
